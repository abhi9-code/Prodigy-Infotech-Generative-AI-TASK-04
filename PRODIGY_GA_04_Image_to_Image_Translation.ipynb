# Task 04 - Image-to-Image Translation using Pix2Pix (cGAN)

## Objective

The objective of this project is to implement an image-to-image translation model using a Conditional Generative Adversarial Network (cGAN) called Pix2Pix. The model learns a mapping between paired input and target images, enabling realistic image translation for tasks such as edge-to-image, sketch-to-photo, and day-to-night conversion.

In this project, a pre-trained Pix2Pix model is used to demonstrate image-to-image translation efficiently without training the network from scratch.
